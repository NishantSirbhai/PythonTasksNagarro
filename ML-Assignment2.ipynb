```python
train=pd.read_csv("Downloads\Assignment 2\Assignment 2\problem_2\problem_2_train.csv")
test=pd.read_csv("Downloads\Assignment 2\Assignment 2\problem_2\problem_2_test.csv")
train.head()
```

```python
test['claimintdate']= pd.to_datetime(test['claimintdate'],dayfirst=True)
test['pickupStartdate']= pd.to_datetime(test['pickupStartdate'],dayfirst=True)
test['mobilepurchasedate']= pd.to_datetime(test['mobilepurchasedate'],dayfirst=True)
test['memcreateddate']= pd.to_datetime(test['memcreateddate'],dayfirst=True)
test['datediff']=test['pickupStartdate']-test['claimintdate']
test['pindiff']=pow(pow(test['custpincode']-test['hubpincode'],2),1/2)
test['datediff']=test['datediff'].dt.days

from sklearn import preprocessing 
  
label_encoder = preprocessing.LabelEncoder() 
   
test['mobmake']= label_encoder.fit_transform(test['mobmake']) 
test['custpincodecategory']=label_encoder.fit_transform(test['custpincodecategory'])
test['claimtype']=label_encoder.fit_transform(test['claimtype'])
test.head()
```

```python
train['claimintdate']=pd.to_datetime(train['claimintdate'],dayfirst=True)
train['pickupStartdate']=pd.to_datetime(train['pickupStartdate'],dayfirst=True)
train['mobilepurchasedate']=pd.to_datetime(train['mobilepurchasedate'],dayfirst=True)
train['memcreateddate']=pd.to_datetime(train['memcreateddate'],dayfirst=True)
train['datediff']=train['pickupStartdate']-train['claimintdate']
train['pindiff']=pow(pow(train['custpincode']-train['hubpincode'],2),1/2)
train['datediff']=train['datediff'].dt.days
```

```python
from sklearn import preprocessing 
  
label_encoder = preprocessing.LabelEncoder() 
   
train['mobmake']= label_encoder.fit_transform(train['mobmake']) 
train['custpincodecategory']=label_encoder.fit_transform(train['custpincodecategory'])
train['claimtype']=label_encoder.fit_transform(train['claimtype'])
train.head()
```

```python
X=train.iloc[:,[0,1,2,5,6,7,9,13,16,17]]
y=train['tat_in_days']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.5,random_state=1)
X
```

# Using Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
y_predict=lr.predict(X_test)
```

```python
np.sqrt(metrics.mean_squared_error(y_test,y_predict))
```
